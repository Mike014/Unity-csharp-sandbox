### Player (Logica Giocatore)

*Script fondamentali per il controllo del personaggio principale.*

| Script | Funzione |
| --- | --- |
| **PlayerController** | Movimento WASD con `Rigidbody`, preserva la gravità (asse Y) e normalizza l'input. |
| **PlayerBehavior** | Movimento avanzato: avanti/indietro + rotazione tramite `MovePosition` e `MoveRotation`. |
| **LifeController** | Sistema salute: gestione danno (Click Sx = -5 HP), eventi `onHealthChanged` e `onPlayerDied`. |

### Other (Comportamenti Vari)

*Utility e logiche secondarie per l'ambiente e gli attori.*

| Script | Funzione |
| --- | --- |
| **CameraBehavior** | Follow camera con offset: usa `LateUpdate`, `LookAt` e gestisce il blocco del cursore. |
| **AnimationBasic** | Semplice rotazione continua di un oggetto sull'asse X (es. monete o pickup). |
| **PlayerJump** | Meccanica di Double Jump: reset dei salti disponibili alla collisione con il tag "Ground". |
| **ItemBehavior** | Logica Collezionabile: l'oggetto si autodistrugge quando collide con il "Player". |
| **EnemyBehavior** | Trigger Zone: rileva e logga l'entrata o l'uscita del Player da un'area. |

### Pattern e Best Practices Utilizzati

* **Comunicazione Disaccoppiata:** Uso di `UnityEvents` per far parlare gli script senza creare dipendenze rigide (es. `LifeController` non conosce la UI, lancia solo un evento).
* **Fisica Corretta:** Utilizzo di `Rigidbody` all'interno di `FixedUpdate` per garantire movimenti fluidi indipendenti dal framerate.
* **Gestione Input:** Implementazione di `Input.GetAxis` per supportare nativamente sia tastiera che controller.
* **Ciclo di Vita:** Utilizzo di `LateUpdate` per la telecamera, assicurando che segua il player solo dopo che quest'ultimo ha terminato il suo movimento.
